```python
column_names = df.isnull().sum().sort_values(ascending = False).index
column_names
find_null_columns = df.isnull().sum().sort_values(ascending = False).values
for col in range(len(find_null_columns)):
    if find_null_columns[col] > 0:
        print (column_names[col])
        df.drop(column_names[col], axis=1, inplace=True)
```

```python
column_names = df.isna().sum().sort_values(ascending = False).index
column_names
find_na_columns = df.isna().sum().sort_values(ascending = False).values
for col in range(len(find_na_columns)):
    if find_na_columns[col] > 0:
        print (column_names[col])
        df.drop(column_names[col], axis=1, inplace=True)
```
